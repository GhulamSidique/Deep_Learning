### here in this notebook we will work on advance regression problem to predict the median value of house 

```python
# importing libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
# for scaling the data
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# for model
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras import metrics
```

```python
# load the data
d_set=pd.read_csv('boston.csv')
d_set.head()
```

```python
# data description
d_set.describe()
```

### result: from above description of the dataset it can be seen that there is huge variations between the values therefore, we need to scale the data

```python
# scalling
scaler=MinMaxScaler()

# fit the data
scaler.fit(d_set)

# create a list of column names
names=[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']]

# transform the data and convert it into a dataframe
d_set=pd.DataFrame(scaler.fit_transform(d_set), columns=names)

d_set.head()
```

```python
# separate the x and y values 
x=d_set.drop(columns=['MEDV'], axis=1)
y=d_set['MEDV']
```

```python
# split the data
xtrain, xtest, ytrain, ytest=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# create the layers for the model
sq=Sequential()

# input and the first hidden layer
sq.add(Dense(input_dim=13, units=20, kernel_initializer='uniform',activation='relu'))

# 2nd hidden layer
sq.add(Dense(units=10, kernel_initializer='uniform', activation='relu'))

# output layer
# since we have regression probleam hence we will make use of linear activation function here
sq.add(Dense(units=1, activation='linear'))
```

```python
# now compilte the model
# loss function for regression problems is MSE
sq.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
# now fit the model
sq.fit(xtrain, ytrain, batch_size=100, epochs=150)
```

```python
sq.evaluate(xtest, ytest), sq.evaluate(xtrain, ytrain)
```

## result: since the loss value is very low as 0.0045 and therefore we can conclude that the model has the highest accuracy score
