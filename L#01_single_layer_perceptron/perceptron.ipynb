# Single layer Perceptron
### this is a single base neural network.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
# for model
from sklearn.linear_model import  Perceptron
```

```python
# load the dataset
d_set=pd.read_csv("placement.csv")
d_set.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
d_set.head()
```

```python
# graph
sns.scatterplot(x=d_set['cgpa'], y=d_set['iq'], data=d_set, hue='placement')
plt.show()
```

```python
# separate x and y data
x=d_set.iloc[:,:-1]
y=d_set['placement']

# split the data
xtrain, xtest, ytrain, ytest=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
# model
pr=Perceptron(alpha=0.0001)
# fit the data
pr.fit(xtrain, ytrain)
```

```python
# scores
pr.score(xtrain, ytrain)*100, pr.score(xtest, ytest)*100
```

```python
# plot decision tree
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=pr)
plt.show()
```

### result: from above graph it is clear that the sinlgle layer perceptron is not performing well
### therefor, we perper multi-layer perceptron
